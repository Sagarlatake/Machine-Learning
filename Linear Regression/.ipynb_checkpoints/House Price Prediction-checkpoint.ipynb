```python
# apply label encoding on required columns
# Import label encoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['Suburb'].head()
```

```python
# use on Suburb column
# fit
le.fit(data['Suburb'])
```

```python
# fit transform
data['Suburb'] = le.fit_transform(data['Suburb'])
```

```python
# check data
data['Suburb'].head()
```

```python
# drop address column
data.drop(columns= 'Address', inplace=True)
data.head()
```

```python
# apply label encoding on all remaining columns
# drop price column
cols = ["Type","Method", "SellerG", "CouncilArea", "Regionname","Date"]


def encode(c):
    le.fit(data[c])
    data[c] = le.fit_transform(data[c])

for c in cols:
    encode(c)
```

```python
# check data
data.head()
```

```python
# split data in X and y (independent features and dependent feature)
X = data.drop(columns='Price')
```

```python
X.head()
```

```python
X.shape
```

```python
y = data['Price'].copy()
y.head()
```

```python
y.shape
```

# # Perform train-test spliting
X - 2 parts
y - 2 parts

xtrain, xtest,
ytrain, ytest.

```python
xtrain, xtest = X[0:10000], X[10000:]
ytrain, ytest = y[0:10000], y[10000:]
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
ytest.shape
```

```python
# Apply model - Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
dir(lr_model)
```

```python
lr_model.fit(xtrain, ytrain)
```

```python
lr_model.intercept_
```

```python
lr_model.coef_
```

```python
# let predict xtest

pred = lr_model.predict(xtest)
```

```python
len(pred)
```

```python
pred
```

```python
pred_df = pd.DataFrame(pred)
```

```python
# Predicted values
pred_df.head()
```

```python
# Actual values
ytest.head()
```

```python
# get error

from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(ytest,pred_df)
print("Error: ",mae)

mse = mean_squared_error(ytest, pred_df)
print("Error:", mse)
```
